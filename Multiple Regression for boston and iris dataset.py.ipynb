```python
# Here we have import some classes like matplotlib,numpy
# Then we have import datasets like boston and linear_model from module named sklearn
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
from sklearn.model_selection import train_test_split
boston=datasets.load_boston()
x=boston.data
y=boston.target
# Here we are splitting the value of x in x_train,x_test and y into y_train,y_test with the help of function named as train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)
# Now here we are defining object named reg where linear_model is a class and LinearRegression is a function
reg=linear_model.LinearRegression()
reg.fit(x_train,y_train) 
print("Regression Coefficients:",reg.coef_)
# Here we are predicting the values of x_train and y_train respectively
plt.scatter(reg.predict(x_train),reg.predict(x_train)-y_train,marker='+',color="black")
plt.scatter(reg.predict(x_test),reg.predict(x_test)-y_test,marker='*',color="red")
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Boston Data')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
from sklearn.model_selection import train_test_split
import pandas as pd
from pandas import DataFrame
iris=datasets.load_iris()
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x_train,y_train,test_size=0.5)
reg=linear_model.LinearRegression()
reg.fit(x_train,y_train)
plt.scatter(reg.predict(x_train),reg.predict(x_train)-y_train,marker='*',color="blue")
plt.scatter(reg.predict(x_test),reg.predict(x_test)-y_test,marker='+',color="red")
plt.xlabel("Sepal Parameters")
plt.ylabel("Features_names")
plt.show()

```
