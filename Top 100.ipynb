```python
import nltk
from nltk.corpus import stopwords
nltk.download_shell()
```

```python
import numpy as np
import pandas as pd
import string
file=open('text.txt','r+')
c=file.read()
c.replace('\n','')

d=[char for char in c.split() if char.lower() not in stopwords.words('english')]
d=' '.join(d)
lists=[char for char in d if d not in string.punctuation]
lists=''.join(lists)
```

```python
lists=lists.replace(",",'')
lists=lists.replace("-",'')
lists=lists.replace("'",'')
lists=lists.replace(".",'')
lists=lists.replace('"','')
lists=lists.replace('_','')
lists=lists.replace("/",'')
lists=lists.replace('*','')
lists=lists.replace('?','')
lists=lists.replace(':','')
lists=lists.replace('(','')
lists=lists.replace(')','')
lists=lists.replace('!','')
lists=lists.replace('”','')
lists=lists.replace('“','')
lists=lists.replace(';','')
lists=lists.replace('&','')

```

```python
print(lists)
```

```python
import nltk
allWords = nltk.tokenize.word_tokenize(lists)

stop_words = set(nltk.corpus.stopwords.words('english'))
allWordExceptStopDist = nltk.FreqDist(w.lower() for w in allWords if w not in stop_words)
allWordExceptStopDist.most_common(100)
```

```python
c=[]
for i in range(0,100):
    c.append(allWordExceptStopDist.most_common(100)[i][0])
```

```python
d=[]
for i in range(0,100):
    d.append(allWordExceptStopDist.most_common(100)[i][1])
print(d)
```

```python


with open('file1.txt', 'w') as filehandle:  
    filehandle.writelines("%s\n" % place for place in c)
with open('file2.txt', 'w') as filehandle1:
    filehandle1.writelines("%d\n" % p for p in d)
```

```python
with open("file1.txt") as xh:
    with open('file2.txt') as yh:
        with open("file3.txt","w") as zh:
      #Read first file
          xlines = xh.readlines()
      #Read second file
          ylines = yh.readlines()
      #Combine content of both lists  and Write to third file
          for line1, line2 in zip(xlines, ylines):
            zh.write("{} : {}\n".format(line1.rstrip(), line2.rstrip()))
```
